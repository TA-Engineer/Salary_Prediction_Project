#### Dropping Rows for columns missing less than 2% of Data
For columns with less than 2% of missing data, rows with null/Nan values will be dropped. This should not effect our prediction from 12497 records

```python
lessthan_2 = Salaries.shape[0]*0.02
print('2% of rows = ',lessthan_2)

cols_null_less_2 = [col for col in Salaries.columns if (Salaries[col].isnull().sum()>0) & (Salaries[col].isnull().sum()<lessthan_2)]
print('Columns with less than 2% Values to be dropped is/are: ',cols_null_less_2)

Salaries.dropna(subset=cols_null_less_2,axis=0, inplace=True)
#Salaries.head(3)
```

```python
print(Salaries.shape)

Salaries.head(2)
```

#### Dropping people who took more than 4 hours to complete the survey
Time from Start to Finish (seconds)' column


```python
minutes = Salaries['Time from Start to Finish (seconds)'].astype(str).astype(int)/60
average_time = minutes.mean()
median_time = minutes.median()
print('mean: ',average_time, ' median: ', median_time)
```

Justification:<br>
We can see that median time to complete the survey is 10.7mins where as the average time to complete the survey is 238.2 mins. This could be because some people maybe just left survey open without completing. We will drop records with time to finish is more than 4 hours = 240mins.
and since half the people took approc 10mins and other half took more than 10mins 240 mins is a good estimation to eliminate rows


```python
Salaries = Salaries[Salaries['Time from Start to Finish (seconds)'].astype(str).astype(int)<(240*60)]
Salaries.shape
```

#### Dropping  - 'Time from Start to Finish (seconds)' column

# Feature Selection

Feature Engineering as a tool in Machine Learning:<br>

Feature engineering/selection is an important tool in Machine learning specially dealing with large datasets as impacts the performance of a ML model.
By applying feature engineering we can reduce dependencies on our target variable. When working with large datasets we have useful as well as irrelevant features, in our case we 234 features which is very high. by the use of feature
engineering we can reduce irrelevant features in our case. This can also help in improving model accuracy.non -relevant relevant features can inverserly impact our model performance by decreasing accuracy. In addition Feature selection process can do the following:
<li>Reduces Overfitting
<li>Improves Accuracy
<li>Reduces Training Time

#### Recursuve Feature Elimination - Feature Selection

Since we have 200+ features and some of them are less relevant predicting the target variable we will use recursive feature elimination process to eliminate some irrelevant features.

We will use the Scikit Learn RFE method for this task. We will use an extrernal estimator to assign weights to the features and then the RFE algorithm will run recuresively
considering smaller and smaler set of features. AS per Scikit Learn: the estimator is trained on the initial set of features and the importance of each 
feature is obtained either through a coef_ attribute or through a feature_importances_ attribute.

```python
def plot_learning_curve(estimator, title, X, y, ylim=None, cv=None, n_jobs=1,\
                        train_sizes=np.linspace(.1, 1.0, 5), scoring='accuracy'):
    
    plt.figure(figsize=(10,6))
    plt.title(title)
    
    if ylim is not None:
        plt.ylim(*ylim)
        
    plt.xlabel("Training examples")
    plt.ylabel(scoring)
    
    train_sizes, train_scores, test_scores = learning_curve(estimator, X, y, cv=cv, scoring=scoring, n_jobs=n_jobs, train_sizes=train_sizes)
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    
    plt.grid()

    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,\
                     train_scores_mean + train_scores_std, alpha=0.1, \
                     color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,\
                     test_scores_mean + test_scores_std, alpha=0.1, color="g")
    
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r",label="Training score")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g", label="Cross-validation score")
    plt.legend(loc="best")
    
    return plt
```

```python
# Calculating Bias and Variance in Data

def bias(y, y_pred):
    y = np.array(y)
    y_pred = np.array(y_pred)
    return (np.mean((y_pred-y)**2))

def variance (y, y_pred):
    y = np.array(y)
    y_pred = np.array(y_pred)
    y_avg = np.mean(y_pred)
    return (np.mean((y_pred-y_avg)**2))
```

```python
def crossv(model1, X1, y1):    
    model = model1
    scaler = StandardScaler()
    kfold = KFold(n_splits=10)
    kfold.get_n_splits(X1)

    accuracy = np.zeros(10)
    np_idx = 0
    outcomes = []
    score_r2=[]
    bias_cv = []
    var_cv = []
    mse = []

    for train_idx, test_idx in kfold.split(X):
        X_train, X_test = X1.values[train_idx], X1.values[test_idx]
        y_train, y_test = y1.values[train_idx], y1.values[test_idx]

        X_train = scaler.fit_transform(X_train)
        X_test = scaler.transform(X_test)

        model.fit(X_train, y_train)

        predictions = model.predict(X_test)
        predict_probability = model.predict_proba(X_test)
        

        TN = confusion_matrix(y_test, predictions)[0][0]
        FP = confusion_matrix(y_test, predictions)[0][1]
        FN = confusion_matrix(y_test, predictions)[1][0]
        TP = confusion_matrix(y_test, predictions)[1][1]
        total = TN + FP + FN + TP
        ACC = (TP + TN) / float(total)

        accuracy[np_idx] = ACC*100
        np_idx += 1

        outcomes.append(accuracy)
        score_r2.append(r2_score(y_test, predictions))
        bias_cv.append(bias(y_test, predictions)) 
        var_cv.append(variance(y_test, predictions)) 
        mse.append(abs(mean_absolute_error(y_test, predictions)))


        print ("Fold {}: Accuracy: {}     Bias: {}   Variance: {}  R2 Score: {}   MSE: {}% ".format(np_idx, round(ACC,3), bias(y_test, predictions), variance(y_test, predictions), r2_score(y_test, predictions), abs(mean_absolute_error(y_test, predictions))))  

    print ("\n Average Score: {}% ({}%)".format(round(np.mean(accuracy),3),round(np.std(accuracy),3)))
    print("\n Variance: {}%".format(round(np.var(outcomes),3)))
    print("\n Standard Deviation: {}%".format(np.std(outcomes)))
    
    plot_learning_curve(model,'Logistic Regression', X1, y1, cv=10)
```
